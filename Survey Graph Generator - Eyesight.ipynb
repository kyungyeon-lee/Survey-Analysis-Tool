```python
#https://www.dataquest.io/blog/how-to-analyze-survey-data-python-beginner/
import pandas as pd
import numpy as np
import ipywidgets as widgets

df = pd.read_csv('test.csv')
full = df[["ResponseId", "Q2", "Q3", "Q4", "Q5", "Q6", "Q7", "Q8", "Q9", "Q10", "Q11", "Q12", "Q13", "Q24", "Q25"]]
```

```python
#Enter cross tab question
Col = "Q3" #Static - eyesight
Row = "Q25"
lowvision_num = 75
blind_num = 76
```

```python
table = full[[Row, Col]]
df = table.reset_index()  # make sure indexes pair with number of rows
group_1 = []
group_2 = []
df_concated = pd.concat([df[Row],df[Col]], axis=1)
df_concated = df_concated.drop(0).drop(1)


for index, row in df_concated.iterrows():
    if row[Col] == '1':#TODO1: Automatically
        group_1.append(row[Row])
    else:
        group_2.append(row[Row])

df_group_1 = pd.Series(group_1)
df_group_1 = df_group_1.str.split(',', expand=True)
df_group_1.stack()
cnt_group_1 = df_group_1.stack().value_counts()

df_group_2 = pd.Series(group_2)
df_group_2 = df_group_2.str.split(',', expand=True)
df_group_2.stack()
cnt_group_2 = df_group_2.stack().value_counts()
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig = go.Figure(data=[go.Bar( #TODO3 Automatically
    x=cnt_group_1.index,
    y=round(100*cnt_group_1/lowvision_num,2),
    text=round(100*cnt_group_1/lowvision_num,2),
    textposition='outside',
    name='Low Vision',
    marker_color='indianred',
), go.Bar(
    x=cnt_group_2.index,
    y=round(100*cnt_group_2/blind_num,2),
    text=round(100*cnt_group_2/blind_num,2),
    textposition='outside',
    name='Blind',
    marker_color='lightsalmon'
 )])
fig.update_layout(
    autosize=False,
    width=1200,
    height=500,
    yaxis_title="Percentage(%)",)
fig.update_layout(yaxis_range=[0,100])
fig.update_xaxes(type='category')
fig.show()
```
